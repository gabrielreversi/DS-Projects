```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/bread basket.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

def encoder(x):
    if x <=0:
        return 0
    if x >= 1:
        return 1

df_ap = df.groupby(['Transaction','Item'])['Item'].count().reset_index(name='Count')
df_bask_ap = df_ap.pivot_table(index='Transaction', columns='Item', values='Count', aggfunc='sum').fillna(0)
df_bask_ap_set = df_bask_ap.applymap(encoder)
df_bask_ap_set.head()
```

```python
freq_item = apriori(df_bask_ap_set, min_support=0.05, use_colnames=True)
freq_item
```

```python
apriori_rules = association_rules(freq_item, metric = 'lift', min_threshold = 0.05)
apriori_rules.sort_values('confidence', ascending = False, inplace = True)
apriori_rules
```

```python
freq_item_1 = apriori(df_bask_ap_set, min_support=0.01, use_colnames=True)
freq_item_1
```

```python
apriori_rules_1 = association_rules(freq_item_1, metric = 'lift', min_threshold = 0.01)
apriori_rules_1.sort_values('confidence', ascending = False, inplace = True)
apriori_rules_1.head()
```

```python
apriori_rules_1.to_csv('apriori_rules_1.csv')
```
